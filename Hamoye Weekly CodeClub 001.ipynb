### Importing Python Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing Dataset

```python
url = 'https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv'
cell_phones = pd.read_csv(url)
cell_phones.head()
```

```python
cell_phones.shape
```

```python
url = 'https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv'
population = pd.read_csv(url)
population.head()
```

```python
population.shape
```

```python
url = 'https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv'
geo_country = pd.read_csv(url)
geo_country.head()
```

```python
geo_country.shape
```

### Joining ‘Cell phones’ and ‘population’ dataframe (With an inner join on "country" & "year")

```python
cell_phones_pop = cell_phones.merge(population, on = ['Country', 'year'], how = 'inner')
```

```python
cell_phones_pop
```

### Renaming the columns(country' to 'geo', and 'name' to 'Country')

```python
geo_country.rename(columns={"country": "geo", "name": "Country"}, inplace= True)
```

### Selecting the 'geo' and 'Country' columns from the 'geo_country' dataframe

```python
gc = geo_country.loc[:, ['geo', 'Country']]
gc
```

### Joining the selected columns(geo and country) from the 'geo_country' dataframe, and join with cell_phones_pop

```python
cell_phones_pop_gc = cell_phones_pop.merge(gc)
```

```python
cell_phones_pop_gc
```

#### To calculate the number of cell phones per person, divide the total number of cell phones by the total number of population

```python
cell_phones_pop_gc['cell_phones_per_person'] = cell_phones_pop_gc['cell_phones_total']/cell_phones_pop_gc['population']
```

```python
cell_phones_pop_gc
```

### Using ".str.upper()" geo column to make the ‘geocodes' uppercase instead of lowercase.

```python
cell_phones_pop_gc['geo'] = cell_phones_pop_gc['geo'] .str.upper()
```

```python
cell_phones_pop_gc['geo']
```

```python
cell_phones_pop_gc
```

#### Using the ".describe()" method to Generate descriptive statistics

```python
cell_phones_pop_gc.describe()
```

### The top 5 countries with the most cell phones total in 2017

```python
cell_phones_subset = cell_phones_pop_gc[cell_phones_pop_gc['year'] == 2017].sort_values('cell_phones_total', ascending = False)
```
